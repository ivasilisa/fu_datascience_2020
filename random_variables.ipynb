```python
import numpy as np
import matplotlib.pylab as plt
```

### Ex 1

```python
s = []
for i in range(1_000_000):
    throw1 = np.random.randint(low=1, high=7, size=1)[0]
    throw2 = np.random.randint(low=1, high=7, size=1)[0]
    s.append(throw1+throw2)
    
```

```python
plt.hist(s, density=True, bins=11)
plt.show()
```

### Ex 2

```python
n, p = 1, 0.4
s = np.random.binomial(n, p, 100)
```

```python
pmf = [np.sum(s)/100, 1-np.sum(s)/100]
plt.bar([0, 1], height=pmf, alpha=0.5, label='sim')
plt.bar([0, 1], height=[0.4, 0.6], alpha=0.5, label='theoretical')

plt.legend()
plt.show()
```

### Ex 3

```python
from scipy import stats
```

```python
mu, sigma = 1, 2 # mean and standard deviation
s = np.random.normal(mu, sigma, 1000)
```

```python
plt.hist(s, density=True)
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
plt.plot(x, stats.norm.pdf(x, mu, sigma))
plt.show()
```
