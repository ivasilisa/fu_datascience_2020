```python
import numpy as np
import matplotlib.pylab as plt
```

### Ex 1

```python
mean = np.array([1,2])
cov = np.array([[0.3,0.2],[0.2,0.5]])
dist = np.random.multivariate_normal(mean,cov,10_000)
```

```python
plt.scatter(x=dist[:,0],y=dist[:,1])
plt.show()
```

```python
np.mean(dist, axis=0)
```

```python
np.std(dist, axis=0)**2
```

```python
plt.hist(dist[:,0],bins=100)
plt.show()
```

```python
plt.hist(dist[:,1],bins=100)
plt.show()
```

### Ex2 

```python
mean = np.array([1,2])
cov = np.array([[0.3,0],[0,0.5]])
dist = np.random.multivariate_normal(mean,cov,10_000) #create multivariate gaussian of two independent random variables
plt.scatter(x=dist[:,0],y=dist[:,1])
plt.show()
```

```python
mu, sigma = 1, 0.3 # mean and standard deviation
X1 = np.random.normal(mu, sigma**0.5, 10000) #create the first random variable
plt.hist(X1,bins=100)
plt.show()
```

```python
mu, sigma = 2, 0.5 # mean and standard deviation
X2 = np.random.normal(mu, sigma**0.5, 10000) #create the first random variable
plt.hist(X2,bins=100)
plt.show()
```

```python
plt.scatter(x=dist[:,0],y=dist[:,1],label="multivariate case",alpha=0.5)
plt.scatter(x=X1,y=X2,label="two distributions",alpha=0.5)
plt.legend()
plt.show()
```

### Ex3 

```python
mean = np.array([1,2])
cov = np.array([[0.3,0.2],[0.2,0.5]])
dist3 = np.random.multivariate_normal(mean,cov,1_000_000) #create multivariate gaussian of two independent random variables
plt.scatter(x=dist3[:,0],y=dist3[:,1])
plt.show()
```

```python
plt.hist(dist3[(dist3[:, 1] > 4.45) & (dist3[:, 1] < 4.55), 0])
plt.hist(dist3[(dist3[:, 1] > -1.05) & (dist3[:, 1] < -0.60), 0])
plt.xlim([-2, 4])
plt.show()
```
